```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
index = ['Sleep', 'Essential Activities', 'Downtime Activities', 'Mental and Physical Health', 'Work']
```

```python
def make_pct(values):
    def my_pct(pct): #autopct will pass in the percent, which we need to convert back to values
        total = sum(values)
        val = int(round(pct*total/100.0))
        return '{p:.1f}%  ({v:d})'.format(p=pct, v=val)
    return my_pct
```

```python
aarondata = [7.3, 2.75, 2.83, 2.72, 8.4]
aaron = pd.Series(aarondata, name='')
aaron.plot.pie(labels=index, autopct=make_pct(aarondata), fontsize=14, figsize=(6,6))
circle = plt.Circle((0,0), 0.75, color='w')
plt.gcf().gca().add_artist(circle)
plt.title("Aaron's Daily Activities", size=22)
plt.show()
```
